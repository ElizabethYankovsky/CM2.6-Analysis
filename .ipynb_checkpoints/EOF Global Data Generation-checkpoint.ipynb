```python
saved_eof1u = u_full[0,:]*np.nan
saved_eof1v = v_full[0,:]*np.nan
saved_pvaru = u_full[0,0,:]*np.nan
saved_pvarv = v_full[0,0,:]*np.nan

for i in range(len(u_full[0,0,:,0])):
    for j in range(len(u_full[0,0,0,:])):
        u = u_full[:,:,i,j]
        v = v_full[:,:,i,j]
        
        if (np.ma.is_masked(u[0,0]))==False:
            solver_u = Eof(u)
            solver_v = Eof(v)
            eofu = solver_u.eofs(neofs=5)
            eofv = solver_v.eofs(neofs=5)

            pvaru = solver_u.varianceFraction(neigs=5)
            pvarv = solver_v.varianceFraction(neigs=5)
            
            saved_eof1u[:,i,j] = eofu[0,:]
            saved_eof1v[:,i,j] = eofv[0,:]
            saved_pvaru[i,j] = pvaru[0]
            saved_pvarv[i,j] = pvarv[0]
    print(i)    

#u    = u_full[:,:,77,237]
#v    = v_full[:,:,77,237]
```
